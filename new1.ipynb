```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt')
import streamlit as st
import plotly.express as px
```

```python
sms_data = [
    "Your order for pizza of Rs 250 is successful",
    "Paid Rs 120 at momos stall",
    "Bought burger for Rs 150",
    "Payment of Rs 350 done at PVR Cinemas",
    "Rs 90 spent on tea",
    "Rs 500 debited for monthly metro pass",
    "Purchased apple and bananas for Rs 200",
    "Dinner bill Rs 450 at food court",
]
categories = {
    "Food": ["pizza", "burger", "momos", "food", "stall", "dinner", "banana", "apple"],
    "Movies": ["pvr", "cinema", "movie", "theatre"],
    "Travel": ["metro", "uber", "ola", "bus", "train"],
    "Miscellaneous": []
}
```

```python
import re
from nltk.tokenize import word_tokenize

def analyze_sms(text):
    text_lower = text.lower()
    words = word_tokenize(text_lower)

    # extract amount
    amount = 0
    amt = re.findall(r"rs[\s\.]*([0-9]+)", text_lower)
    if amt:
        amount = int(amt[0])

    # classify category using keyword matching
    final_category = "Miscellaneous"
    for cat, keys in categories.items():
        if any(key in text_lower for key in keys):
            final_category = cat
            break

    return final_category, amount
```

```python
results = []
for msg in sms_data:
    category, amount = analyze_sms(msg)
    results.append([msg, category, amount])

df = pd.DataFrame(results, columns=["Message", "Category", "Amount"])
df
summary = df.groupby("Category")["Amount"].sum().reset_index()
summary
```

```python
plt.figure(figsize=(10,5))
plt.bar(summary["Category"], summary["Amount"])
plt.xlabel("Category")
plt.ylabel("Total Monthly Spend (Rs)")
plt.title("Your Monthly Spending Overview")
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

```
